```
import pandas as pd
import numpy as np
import gensim
import gzip
import nltk
```

```
df=pd.read_json("http://snap.stanford.edu/data/amazon/productGraph/categoryFiles/reviews_Office_Products_5.json.gz",lines=True)
df=df.iloc[0:500] 
```

```
from gensim.models import Word2Vec
from nltk import word_tokenize
nltk.download('stopwords')
from nltk.corpus import stopwords
stop_words=set(stopwords.words('english'))
```

```
nltk.download('punkt')
```

```
punctuations = '''!()-[]{};:'"\,<>./?@#$%^&*_~''' #Function to remove punctuations
def punc(str):
  no_punct = ""
  for char in str:
    if char not in punctuations:
      no_punct = no_punct + char
  return no_punct
```

```
def stop_w(text): #Tokenizing text and removing the stop words
  text=text.lower()
  tokens=word_tokenize(text)
  nostop=[words for words in tokens if not words in stop_words]
  return nostop
```

```
df['reviewText']=df['reviewText'].apply(punc)
```

```
df['reviewText_nostop']=df['reviewText'].apply(stop_w)
```

```
def uniq_words(text):
  uniq=[]
  for words in text:
    if words not in uniq:
      uniq.append(words)
  return uniq
```

```
df['reviewText_nostop']=df['reviewText_nostop'].apply(uniq_words)
```

```
df.head() #Displaying dataframe with new column added
```

```
wordlist=[[word] for word in df['reviewText_nostop'][0]]
```

```
model=Word2Vec(wordlist, size=100, window=5, min_count=1, workers=4)
```

```
for i in wordlist:
  for j in wordlist:
    if j>i:
      print('Similarity between ',i[0],' & ',j[0],' is: ',model.similarity(i[0],j[0]))
    else:
      continue
```
